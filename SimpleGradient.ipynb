```python
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 3*x**2 - 6*x + 8
def grad_f(x):
    return 6*x-6
```

```python
def gradient_descent(x0, eta, iterations):
    x = x0
    x_values = [x]
    f_values = [f(x)]
    
    for i in range(iterations):
        x = x - eta * grad_f(x)
        x_values.append(x)
        f_values.append(f(x))
    
    return x_values, f_values
```

```python
etas = [0.01, 0.1]
iterations = 20

results = {}

for eta in etas:
    x_vals, f_vals = gradient_descent(0, eta, iterations)
    results[eta] = (x_vals, f_vals)
```

```python
plt.figure()

for eta in results.keys():   # safer than using etas
    plt.plot(results[eta][1], label=f"eta={eta}")

plt.xlabel("Iteration")
plt.ylabel("Function Value")
plt.legend()
plt.title("Gradient Descent Convergence")
plt.show()
```

```python
def f2(x):
    return x**4 - 3*x**3 + 2

def grad_f2(x):
    return 4*x**3 - 9*x**2
```

```python
def gradient_descent2(x0, eta, iterations):
    x = x0
    x_values = [x]
    
    for i in range(iterations):
        x = x - eta * grad_f2(x)
        x_values.append(x)
    
    return x_values
```

```python
eta = 0.01
iterations = 100

x_vals1 = gradient_descent2(-1, eta, iterations)
x_vals2 = gradient_descent2(2.5, eta, iterations)

print("Final value starting from -1:", x_vals1[-1])
print("Final value starting from 2.5:", x_vals2[-1])
```
